```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
root = ET.parse('rvko_2023_1.xml')
```

```python
notations = []
descriptions = []

for first_level_node in root.findall("node"):
    notations.append(first_level_node.get("notation"))
    descriptions.append(first_level_node.get("benennung"))
    second_level_elements = first_level_node.find("children")
    for second_level_node in second_level_elements.findall("node"):
        notations.append(second_level_node.get("notation"))
        descriptions.append(second_level_node.get("benennung"))
        third_level_elements = second_level_node.find("children")
        # Third level if available.
        try: 
            for third_level_node in third_level_elements.findall("node"):
                notations.append(third_level_node.get("notation"))
                descriptions.append(third_level_node.get("benennung"))
        except:
            pass
```

```python
df = pd.DataFrame({"notation": notations, "description": descriptions})
```

```python
df.to_csv("rvk_classification_short_flattened.tsv", sep="\t", index=False)
```
